```python
import torch
from diffusers import StableDiffusionPipeline
from transformers import CLIPTokenizer, CLIPTextModel
import gradio as gr
```

```python
# Load the Stable Diffusion Pipeline
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

print("Stable Diffusion Pipeline loaded!")
```

```python
# Define a prompt for image generation
prompt = "A fantasy landscape with castles and dragons flying in the sky."

# Generate the image
generated_images = pipe(prompt, num_inference_steps=50, guidance_scale=7.5)

# Display the generated image
image = generated_images["images"][0]
image.show()

# Save the image locally
image.save("generated_image.png")
print("Image saved as 'generated_image.png'")
```

```python
# Define a function to generate images from a text prompt
def generate_image(prompt):
    result = pipe(prompt, num_inference_steps=50, guidance_scale=7.5)
    return result["images"][0]

# Create a Gradio interface
interface = gr.Interface(
    fn=generate_image,
    inputs="text",
    outputs="image",
    title="Text-to-Image Generator",
    description="Enter a description to generate a corresponding image using Stable Diffusion.",
)

# Launch the interface
interface.launch()
```

```python
# Load the CLIP tokenizer and text model
tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")
text_model = CLIPTextModel.from_pretrained("openai/clip-vit-base-patch32")

# Tokenize the text prompt
inputs = tokenizer("A cute cat sitting on a mountain under a rainbow", return_tensors="pt")

# Generate text embeddings
text_embeddings = text_model(**inputs).last_hidden_state
print("Text Embeddings Shape:", text_embeddings.shape)
```
